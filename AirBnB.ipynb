```python
labels = users_df['country_destination']
users_df.drop('country_destination',axis=1,inplace=True)
tuser_id = test_users_df['id']
train_idx = users_df.shape[0]
```

```python
users_df_ohe = pd.concat((users_df, test_users_df), axis=0, ignore_index=True)
#users_df_ohe.gender.replace('-unknown-', np.nan, inplace=True)
ohe_feats = ['gender', 'signup_method', 'signup_flow', 'language', 'affiliate_channel', 'affiliate_provider', 'first_affiliate_tracked', 'signup_app', 'first_device_type', 'first_browser']
for f in ohe_feats:
    df_all_dummy = pd.get_dummies(users_df_ohe[f], prefix=f)
    users_df_ohe = users_df_ohe.drop([f], axis=1)
    users_df_ohe = pd.concat((users_df_ohe, df_all_dummy), axis=1)
```

```python
dac_dt = pd.to_datetime(users_df_ohe.date_account_created)
users_df_ohe['dac_year'] =dac_dt.dt.year
users_df_ohe['dac_month']=dac_dt.dt.month
users_df_ohe['dac_day'] =dac_dt.dt.day

tfa_dt =users_df_ohe.timestamp_first_active.astype(str)
users_df_ohe['tfa_year'] =tfa_dt.str.slice(0,4,1).astype(int)
users_df_ohe['tfa_month']=tfa_dt.str.slice(4,6,1).astype(int)
users_df_ohe['tfa_day'] =tfa_dt.str.slice(6,8,1).astype(int)
#users_df_ohe['tfa_hour'] = tfa_dt.str.slice(8,10,1).astype(int)
#users_df_ohe['tfa_mins'] = tfa_dt.str.slice(10,12,1).astype(int)
#users_df_ohe['tfa_secs'] = tfa_dt.str.slice(12,14,1).astype(int)

users_df_ohe.drop('date_account_created',axis=1,inplace=True)
users_df_ohe.drop(['id', 'date_first_booking'],inplace=True, axis=1)
users_df_ohe = users_df_ohe.fillna(-1)
av = users_df_ohe.age.values
users_df_ohe['age'] = np.where(np.logical_or(av<14, av>90), -1, av)
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn import svm,grid_search
from sklearn.ensemble import ExtraTreesClassifier
```

```python
user_train = users_df_ohe[:train_idx]
user_test = users_df_ohe[train_idx:]
ftrain,ftest,ltrain,ltest=train_test_split(user_train,labels,test_size=.20,random_state=34)
scaler = MinMaxScaler()
#scaler.fit(user_train)  
#user_train = scaler.transform(user_train)
scaler.fit(ftrain)  
ftrain = scaler.transform(ftrain)
```

```python
clf = ExtraTreesClassifier(criterion='entropy',min_samples_leaf=3,n_estimators=100)

clf.fit(ftrain,ltrain)
```

```python
#ftrain = scaler.transform(ftrain)
clf.score(ftrain,ltrain)
```

```python
#ftest = scaler.transform(ftest)
clf.score(ftest,ltest)
```

```python
user_test = scaler.transform(user_test)
rte_t =clf.predict_proba(user_test)
pred_df = pd.DataFrame(rte_t,columns=clf.classes_)
```

```python
l_start = 0
l_end = pred_df.shape[0]
result_df = pd.DataFrame(columns=['id','index'])

for i in range(l_end):
    vv=pred_df[i:i+1].T
    vvv=vv.sort_values([i],ascending=False)[:5]
    vvv['id']=tuser_id[i]
    v4=vvv.reset_index()[['id','index']]
    result_df=pd.concat([result_df,v4])
result_df.columns=['id','country']
```

```python
result_df.to_csv('/home/mex/DS/Kaggle/AirBnB/submission7.csv',index=False)
```
