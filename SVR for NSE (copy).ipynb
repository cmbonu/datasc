# Data Wrangling

### Convert Date Strings to Date Objects for easier manipulation

```python
price_hist['prc_date']=pd.to_datetime(price_hist['Date'])
price_hist.drop(['Date'],inplace=True,axis=1)

##crude_oil_hist['prc_date']=pd.to_datetime(crude_oil_hist['Date'])
crude_oil_hist['prc_date']=[datetime.strptime(i,'%d/%m/%Y') for i in crude_oil_hist['Date']]
crude_oil_hist.drop(['Date'],inplace=True,axis=1)

exch_rate_hist['prc_date']=pd.to_datetime(exch_rate_hist['Rate Date'])
exch_rate_hist.drop(['Rate Date'],inplace=True,axis=1)
```

### Generate reference array of dates

```python
start_date = datetime(2000,1,1)
end_date = datetime(2017,1,1)
day_count = (end_date-start_date).days
date_array = []
for ticker in price_hist['Ticker'].unique():
    for i in range(day_count):
        newdelta = timedelta(days=i)
        ticker_date = (ticker,start_date+newdelta)
        date_array.append(ticker_date)
dates_frame = pd.DataFrame(date_array,columns=['Ticker','prc_date'])
```

### Merge data frames

```python
crude_oil_hist['prev_prc_date'] = crude_oil_hist['prc_date']+ timedelta(days=1)
exch_rate_hist['prev_prc_date'] = exch_rate_hist['prc_date']+ timedelta(days=1)
price_hist['prev_prc_date'] = price_hist['prc_date']+ timedelta(days=1)

tmp_date_price = pd.merge(dates_frame,price_hist,how='left')
tmp_date_price = pd.merge(tmp_date_price,price_hist[['prev_prc_date','Ticker','ClosePrice']],how='left',
                          left_on=['prc_date','Ticker'],right_on=['prev_prc_date','Ticker'],suffixes=('', '_prev'))

tmp_crude = pd.merge(tmp_date_price,crude_oil_hist[['Price(USD)','prev_prc_date']],how='left',left_on='prc_date',
                     right_on='prev_prc_date')

data_merged = pd.merge(tmp_crude,exch_rate_hist[['BuyRate', 'SellRate','prev_prc_date']],how='left',
                       left_on='prc_date',right_on='prev_prc_date')

data_merged.drop(['prev_prc_date_x','prev_prc_date_y','prev_prc_date_prev','prev_prc_date'],inplace=True,axis=1)

data_merged_day2 = data_merged[['Ticker', 'ClosePrice','Price(USD)', 'BuyRate','SellRate', 'prc_date']]
data_merged_day2['prev_prc_date']=data_merged_day2['prc_date']+ timedelta(days=2)
data_merged_day2.drop(['prc_date'],inplace=True,axis=1)
data_merged = pd.merge(data_merged,data_merged_day2,how='left',left_on=['prc_date','Ticker'],
                       right_on=['prev_prc_date','Ticker'],suffixes=('', '_prev2'))

data_merged.drop(['prev_prc_date'],inplace=True,axis=1)
```

### Filter from Jan 2013 to Dec 2014

```python
data_start_date = datetime(2013,1,1)
data_end_date = datetime(2014,12,31)
data_filtered = data_merged[(data_merged.prc_date>=data_start_date) & (data_merged.prc_date<=data_end_date)]
```

### Fill Missing Values
<ul>
<li>Forward fill is initially used to propagate previous day values to the next day across all features
<li>Backward fill is then applied to fill missing values for the first two days if 2013
</ul>

```python
data_filled = data_filtered.fillna(method='bfill').fillna(method='ffill')
```

### Additional Data wrangling and Feature Engineering
<ul>
<li>Perform One hot encoding for Ticker Symbols.
<li>Split dates into Year, date and day fields.
<li>Expand feature space by cross multiplyting across features.
<li>Split data into train and test sets. (3:1) split by dates
</ul>

```python
data_start_date = datetime(2013,1,1)
data_end_date = datetime(2014,6,30)
train_dataframe=data_filled[(data_filled.prc_date>=data_start_date) & (data_filled.prc_date<=data_end_date)]
test_dataframe=data_filled[(data_filled.prc_date>data_end_date)]
```

```python
rs = ShuffleSplit(data_filled.shape[0], n_iter=1,test_size=.25, random_state=66)
tridx = []
tsidx =[]
for train_index, test_index in rs:
    tridx=train_index
    tsidx=test_index
train_dataframe= data_filled.iloc[tridx]
test_dataframe= data_filled.iloc[tsidx]
```

```python
close_prices = train_dataframe['ClosePrice']
train_dataframe.drop(['ClosePrice'],axis=1,inplace=True)
```

```python
tmp_df=pd.get_dummies(train_dataframe['Ticker'],prefix='Ticker')
train_dataframe = pd.concat((train_dataframe,tmp_df),axis=1).drop(['Ticker'],axis=1)
```

```python
train_dataframe['prc_year']=train_dataframe.prc_date.dt.year
train_dataframe['prc_month']=train_dataframe.prc_date.dt.month
train_dataframe['prc_day']=train_dataframe.prc_date.dt.day
train_dataframe.drop(['prc_date'],axis=1,inplace=True)
```

```python
cols=train_dataframe.columns
```

```python
logcols = train_dataframe.columns
for i in range(0,logcols.size-1):
    for ii in range(i+1,logcols.size):
        train_dataframe[logcols[i]+'_'+logcols[ii]] = train_dataframe[logcols[i]]/train_dataframe[logcols[ii]].replace(0,1)
```

# Model Training and Assesment

<b>A Min-Max scaler is applied to all the features to reduces the influence of high value features.</b>

```python
#ftrain,ftest,ltrain,ltest=train_test_split(data_filled,close_prices,test_size=.10,random_state=16)
scaler = MinMaxScaler()
scaler.fit(train_dataframe)  
ftrain = scaler.transform(train_dataframe)
```

### Instantiate Support Vector Regressor and fit model to training data

```python
#parameters = {'C':[x for x in range(10,20)],'gamma':[x*0.01 for x in range(0,100)]}
parameters = {'C':[x for x in range(10,100)]}
#clf = svm.SVR()
svmr = svm.SVR()
clf = grid_search.GridSearchCV(svmr, parameters)
clf.fit(ftrain, close_prices) 
```

```python
#clf.best_estimator_ 
clf=svm.SVR(C=62, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma=0.0,kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
clf.fit(ftrain, close_prices) 
#SVR(C=10, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma=0.0,
 # kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
```

### Instantiate Artificial Neural Network and fit model to training data

```python
train_dataframe.shape
```

```python
idim =train_dataframe.shape[1]
model = Sequential()
model.add(Dense(10, input_dim=idim))
model.add(Activation('tanh'))
model.add(Dense(1,input_dim=66))
model.compile(loss='mean_absolute_error', optimizer='rmsprop')
```

```python
model.fit(ftrain, close_prices,nb_epoch=500,show_accuracy=True)
```

### Munge test data to training template

```python
def process_datarow(data_merged,cols):
    darray=[]
    for x in cols:
        try:
            darray.append(data_merged[x].iloc[0])
        except:
            if 'Ticker_'+data_merged['Ticker'].iloc[0]==x:
                darray.append(1)
            else:
                darray.append(0)
    return darray
```

```python
test_dataframe['prc_year']=test_dataframe.prc_date.dt.year
test_dataframe['prc_month']=test_dataframe.prc_date.dt.month
test_dataframe['prc_day']=test_dataframe.prc_date.dt.day
#data_mr.drop(['prc_date'],axis=1,inplace=True)
```

```python
#cols =train_dataframe.columns
data_processed = pd.DataFrame(columns=cols)
for i in range(0,test_dataframe.shape[0]):
    data_processed.loc[len(data_processed)]=process_datarow(test_dataframe[i:i+1],cols)
    
logcols = data_processed.columns
for i in range(0,logcols.size-1):
    for ii in range(i+1,logcols.size):
        data_processed[logcols[i]+'_'+logcols[ii]] = data_processed[logcols[i]]/data_processed[logcols[ii]].replace(0,1)

data_scaled = scaler.transform(data_processed)
```
